### 2. How many cars have mpg  35?

```python
import pandas as pd

# Re-load the original CSV to ensure all columns are available
df_original = pd.read_csv('/content/Cars.csv')

# Display the head to confirm it loaded correctly
print(df_original.head())
```

### 3. What is the most common origin for cars with horsepower > 100 and weight < 3000?

```python
# Filter for cars with horsepower > 100
filtered_cars_hp = df_original[df_original['horsepower'] > 100]

# Further filter for cars with weight < 3000
filtered_cars_hp_weight = filtered_cars_hp[filtered_cars_hp['weight'] < 3000]

# Get the value counts for the 'origin' column in the filtered DataFrame
origin_counts = filtered_cars_hp_weight['origin'].value_counts()

# Get the most common origin
most_common_origin = origin_counts.index[0]

print(f"The most common origin for cars with horsepower > 100 and weight < 3000 is: {most_common_origin}")
```

### 4. What is the mean acceleration of cars from Japan? (rounded to 2 decimals)

```python
# Filter for cars from Japan
cars_from_japan = df_original[df_original['origin'] == 'japan']

# Calculate the mean acceleration
mean_acceleration_japan = cars_from_japan['acceleration'].mean()

# Round to 2 decimal places
rounded_mean_acceleration = round(mean_acceleration_japan, 2)

print(f"The mean acceleration of cars from Japan is: {rounded_mean_acceleration}")
```

### 5. Which year had the highest average mpg?

```python
# Group by 'model_year' and calculate the mean of 'mpg' for each year
average_mpg_by_year = df_original.groupby('model_year')['mpg'].mean()

# Find the year with the highest average mpg
year_highest_avg_mpg = average_mpg_by_year.idxmax()
highest_avg_mpg_value = average_mpg_by_year.max()

print(f"The year with the highest average mpg is: {year_highest_avg_mpg}")
print(f"The highest average mpg value is: {highest_avg_mpg_value:.2f}")
```

### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
# 1. Calculate the median mpg
median_mpg = df_original['mpg'].median()

# 2. Filter for cars with mpg above the median
cars_above_median_mpg = df_original[df_original['mpg'] > median_mpg].copy() # .copy() to avoid SettingWithCopyWarning

# 3. Calculate the 'horsepower to weight' ratio for these filtered cars
# Handle potential division by zero or NaN horsepower if not already handled
# As seen in df.info(), 'horsepower' has NaNs, so fill them for this calculation if needed, or drop rows.
# For this problem, I'll assume we want to exclude rows with NaN horsepower for this ratio calculation.
cars_above_median_mpg.dropna(subset=['horsepower', 'weight'], inplace=True) # Drop rows with NaN in these columns

# Ensure 'weight' is not zero to prevent division by zero errors
cars_above_median_mpg = cars_above_median_mpg[cars_above_median_mpg['weight'] != 0]

cars_above_median_mpg['hp_to_weight_ratio'] = cars_above_median_mpg['horsepower'] / cars_above_median_mpg['weight']

# 4. Find the car(s) with the highest 'horsepower to weight' ratio
max_ratio = cars_above_median_mpg['hp_to_weight_ratio'].max()
cars_with_best_ratio = cars_above_median_mpg[cars_above_median_mpg['hp_to_weight_ratio'] == max_ratio]

# 5. Get the name(s) of these car(s)
best_ratio_car_names = cars_with_best_ratio['name'].tolist()

print(f"The median MPG is: {median_mpg}")
print(f"The car(s) with the best horsepower to weight ratio among cars with above-median mpg are: {best_ratio_car_names}")
print(f"The best ratio (horsepower/weight) is: {max_ratio:.4f}")
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
# Group by 'model_year' and 'origin', then calculate the mean 'mpg'
average_mpg_by_year_origin = df_original.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Create the multi-line plot using Seaborn
plt.figure(figsize=(10, 6))
sns.lineplot(data=average_mpg_by_year_origin, x='model_year', y='mpg', hue='origin', marker='o')

# Add title and labels
plt.title('Evolution of Average MPG Over Years by Origin')
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.xticks(average_mpg_by_year_origin['model_year'].unique()) # Ensure all years are shown as ticks
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(title='Origin')
plt.tight_layout()
plt.show()
```

### We define a consistent car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
# 1. Identify car names that appear in more than one model_year.
# Group by 'name' and count unique 'model_year' values for each car.
car_year_counts = df_original.groupby('name')['model_year'].nunique()

# Filter for car names that appear in more than one model_year
multi_year_cars = car_year_counts[car_year_counts > 1].index

# 2. For each such name, compute the standard deviation of mpg across years.
# Filter the original DataFrame to include only these multi-year cars
filtered_df = df_original[df_original['name'].isin(multi_year_cars)]

# Group by 'name' and compute the standard deviation of 'mpg'
mpg_std_by_name = filtered_df.groupby('name')['mpg'].std().dropna() # dropna to handle single appearance stdev=NaN

# 3. Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.
consistent_cars_std = mpg_std_by_name[mpg_std_by_name < 1.0]

if consistent_cars_std.empty:
    print("No consistent car models found with standard deviation < 1.0.")
else:
    # 4. Report the model name(s), number of appearances, and the average mpg.
    # Get the names of the consistent cars
    consistent_car_names = consistent_cars_std.index

    # Calculate number of appearances and average mpg for these consistent cars
    final_report_data = []
    for car_name in consistent_car_names:
        car_data = df_original[df_original['name'] == car_name]
        num_appearances = car_data['model_year'].nunique() # Number of unique years produced
        avg_mpg = car_data['mpg'].mean()
        std_mpg = car_data['mpg'].std() # Re-confirm std for printing
        final_report_data.append({
            'Model Name': car_name,
            'Number of Appearances': num_appearances,
            'Average MPG': avg_mpg,
            'Standard Deviation of MPG': std_mpg
        })

    # Convert to DataFrame for easier sorting and display
    consistent_cars_df = pd.DataFrame(final_report_data)

    # Bonus: Sort the result by number of appearances (descending), then mpg (descending).
    consistent_cars_df_sorted = consistent_cars_df.sort_values(
        by=['Number of Appearances', 'Average MPG'],
        ascending=[False, False]
    ).reset_index(drop=True)

    print("\nConsistent Car Models (standard deviation of mpg < 1.0):")
    print(consistent_cars_df_sorted.to_string(index=False)) # Use to_string to prevent truncation
```
